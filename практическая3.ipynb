### Задание 1

```python
import math

x = 4.4  # начальное значение
eps = 0.001

while True:
    x_new = math.sqrt(20 - 2 * 2**(-x))
    if abs(x_new - x) < eps:
        break
    x = x_new

print(f'Корень: {x:.4f}')
print(f'Проверка: 10 - 0.5*{x:.4f}^2 = {10 - 0.5*x**2:.4f}')
print(f'Проверка: 2^(-{x:.4f}) = {2**(-x):.4f}')
```
