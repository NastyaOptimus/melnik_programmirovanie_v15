```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

### Задание 1

```python
print("Графический метод")

# Строим графики двух функций
x = np.linspace(0, 6, 200)  # можно взять от 0 до 6
y1 = 10 - 0.5 * x**2         # левая часть уравнения
y2 = 2**(-x)                 # правая часть уравнения

plt.plot(x, y1, label='10 - 0.5x²')
plt.plot(x, y2, label='2⁻ˣ')
plt.grid(True)
plt.legend()
plt.title('Отделение корней уравнения 10 - 0.5x² = 2⁻ˣ')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()

# Находим интервалы, где функции пересекаются (разность меняет знак)
print("\nИнтервалы, содержащие корни:")
for i in range(len(x)-1):
    if (y1[i] - y2[i]) * (y1[i+1] - y2[i+1]) < 0:
        print(f"Корень на [{x[i]:.2f}; {x[i+1]:.2f}]")
```

### Задание 2

```python
def f(x):
    return 10 - 0.5 * x**2 - 2**(-x)

a, b = 4.43, 4.46  # интервал из графического метода
eps = 0.001

print('n     a         b         x         f(x)      |x_n - x_{n-1}|')
print('-' * 65)

x_prev = a
n = 1
while True:
    x = a - f(a) * (b - a) / (f(b) - f(a))
    diff = abs(x - x_prev)
    
    print(f'{n:2d}   {a:.6f} {b:.6f} {x:.6f} {f(x):.6f} {diff:.6f}')
    
    if abs(f(x)) < eps or diff < eps:
        print(f'\nКорень: x = {x:.6f}')
        break
        
    if f(a) * f(x) < 0:
        b = x
    else:
        a = x
    
    x_prev = x
    n += 1

print(f'Проверка: 10 - 0.5*{x:.6f}^2 = {10 - 0.5*x**2:.6f}')
print(f'Проверка: 2^(-{x:.6f}) = {2**(-x):.6f}')
```

### Задание 3

```python
def f(x):
    return 10 - 0.5 * x**2 - 2**(-x)

def f_derivative(x):
    # Производная: -x + 2^(-x) * ln(2)
    return -x + 2**(-x) * math.log(2)

x = 4.4  # начальное приближение
eps = 0.001

print('n     x_n        f(x_n)      |x_n - x_{n-1}|')
print('-' * 45)

n = 1
while True:
    x_new = x - f(x) / f_derivative(x)
    diff = abs(x_new - x)
    
    print(f'{n:2d}   {x:.6f}   {f(x):.6f}   {diff:.6f}')
    
    if abs(f(x_new)) < eps or diff < eps:
        print(f'\nКорень: x = {x_new:.6f}')
        break
    
    x = x_new
    n += 1

print(f'Проверка: 10 - 0.5*{x_new:.6f}^2 = {10 - 0.5*x_new**2:.6f}')
print(f'Проверка: 2^(-{x_new:.6f}) = {2**(-x_new):.6f}')
```
